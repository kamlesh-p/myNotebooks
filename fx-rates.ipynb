### Data downloaded from kaggle.com

```python
%matplotlib inline
from matplotlib import pyplot as plt

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
 

import pandas as pd

df = pd.read_csv('fx-rates-per-dollar-20002019/Foreign_Exchange_Rates.csv', index_col=0, usecols=[0, 1, 10], dtype={'INDIA - INDIAN RUPEE/US$':'float'}, parse_dates=[1], na_values='ND')
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.ylabel('INR')
plt.xlabel('INR')
plt.title('INR / 1 US$');
plt.rcParams['figure.figsize'] = [22,9]

plt.hlines(y=df['INDIA - INDIAN RUPEE/US$'].min(), xmin=df['Time Serie'][0], xmax=df['Time Serie'][df[['INDIA - INDIAN RUPEE/US$']].idxmin()], color='g', linestyle='-', label='Min')
plt.vlines(x=df['Time Serie'][df[['INDIA - INDIAN RUPEE/US$']].idxmin()], ymin=df['INDIA - INDIAN RUPEE/US$'].min(), ymax=df['INDIA - INDIAN RUPEE/US$'].min(), color='g', linestyle='-', label='Min')

plt.hlines(y=df['INDIA - INDIAN RUPEE/US$'].max(),xmin=df['Time Serie'][0], xmax=df['Time Serie'][df[['INDIA - INDIAN RUPEE/US$']].idxmax()], color='r', linestyle='-', label="Max")
plt.vlines(x=df['Time Serie'][df[['INDIA - INDIAN RUPEE/US$']].idxmax()], ymin=, ymax=df['INDIA - INDIAN RUPEE/US$'].max(), color='r', linestyle='-', label='Min')

plt.plot(df['Time Serie'], df['INDIA - INDIAN RUPEE/US$'], color='tab:blue')
```
