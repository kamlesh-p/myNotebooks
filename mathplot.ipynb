```python
import pandas as pd
import math
```

```python
df = pd.DataFrame([math.log10(i) for i in range(1, 10000)])
df.plot()
```

```python
df2 = pd.DataFrame(data = [math.log2(i) for i in range(1, 100000)])
df2.plot()
```

```python
df2 = pd.DataFrame(data = [math.log2(i) for i in range(1, 1000000)])
df2.plot()
    
```

```python
df = pd.DataFrame([i*i for i in range(1, 1000)])
df.plot()
```

```python
df = pd.DataFrame([i**i for i in range(1, 10)])
df.plot()
```

```python
df = pd.DataFrame([math.sin(math.radians(i)) for i in range(0, 360)])
df.plot()
```

```python
df = pd.DataFrame([math.cos(math.radians(i)) for i in range(0, 360)])
df.plot()
```

```python
df = pd.DataFrame([math.tan(math.radians(i)) for i in range(0, 360)])
df.plot()
```

```python
df = pd.DataFrame([math.sin(math.radians(i))/math.radians(i) if i !=0 else 1 for i in range(-1080, 1080)])
df.plot()
```

```python
df = pd.DataFrame([math.cos(math.radians(i))/i if i !=0 else 1 for i in range(-360, 360)])
df.plot()
```

```python
df = pd.DataFrame([math.sin(math.radians(i)) + math.cos(math.radians(i+90)) for i in range(0, 90)])
df.plot()
```

```python
df = pd.DataFrame([math.tan(math.radians(i)) for i in range(-60, 60)])
df.plot()
```
