```python
%matplotlib inline
from matplotlib import pyplot as plt

import pandas as pd

sachin_odi_data = pd.read_csv('sachin_odi_scr.csv', dtype={'run':int})
kohli_odi_data = pd.read_csv('kohli_odi_scr.csv', dtype={'run':int})

sachin_dob = '24 Apr 1973'
kohli_dob = '05 Nov 1988'
```

```python
sachin_odi_data
```

```python
%%time
plt.scatter(sachin_odi_data.index, sachin_odi_data['run'][::-1], color='c')
```

```python
%%time
plt.scatter(kohli_odi_data.index, kohli_odi_data['run'][::-1], color='c')
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
fig, ax1 = plt.subplots()

x = sachin_odi_data.index
s_score = sachin_odi_data['run'][::-1]
s_date = pd.to_datetime(sachin_odi_data['date'][::-1], format='%d %B %Y', errors='ignore')


color = 'tab:cyan'
ax1.set_xlabel('Date')
ax1.set_ylabel('score')
ax1.plot(s_date, s_score, color=color)
#ax1.tick_params(axis='y', labelcolor=color)

  # instantiate a second axes that shares the same x-axis

x = kohli_odi_data.index
k_score = pd.to_numeric(kohli_odi_data['run'][::-1])
k_date = pd.to_datetime(kohli_odi_data['date'][::-1], format='%d %B %Y', errors='ignore')

color = 'tab:red'
ax1.plot(k_date, k_score, color=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.set_size_inches(20, 8)
plt.show()
```

```python
fig, ax1 = plt.subplots()

x = sachin_odi_data.index
s_score = sachin_odi_data['run'][::-1]
s_balls = sachin_odi_data['ball'][::-1]
s_date = pd.to_datetime(sachin_odi_data['date'][::-1], format='%d %B %Y', errors='ignore')


color = 'tab:cyan'
ax1.set_xlabel('Date')
ax1.set_ylabel('score')
ax1.bar(x, (s_score/s_balls * 100), color=color)
#ax1.tick_params(axis='y', labelcolor=color)

  # instantiate a second axes that shares the same x-axis

x = kohli_odi_data.index
k_score = kohli_odi_data['run'][::-1]
k_balls = kohli_odi_data['ball'][::-1]
k_date = pd.to_datetime(kohli_odi_data['date'][::-1], format='%d %B %Y', errors='ignore')

color = 'tab:red'
ax1.bar(x, (k_score/k_balls * 100), color=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.set_size_inches(16, 8)
plt.show()
```

```python
x = sachin_odi_data.index
s_score = pd.to_numeric(sachin_odi_data['run'][::-1])
s_balls = pd.to_numeric(sachin_odi_data['ball'][::-1])
s_date = pd.to_datetime(sachin_odi_data['date'][::-1], format='%d %B %Y', errors='ignore')

plt.ylabel('SCORE')
plt.title('Sachin ODI score');

plt.plot(s_date, s_score, color='tab:blue')
plt.savefig('sachin.jpg')
plt.show()

plt.plot(s_date, s_score/s_balls * 100, color='tab:green')
plt.show()
```
